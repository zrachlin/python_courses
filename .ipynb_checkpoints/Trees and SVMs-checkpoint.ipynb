## Getting Started

<a id="import"></a>
### Importing Data

Let's begin by importing the data. For this class, I'll be using a dataset from the fascinating alien world of psychology. To read about it, I encourage you to go [here](http://journal.frontiersin.org/article/10.3389/fpsyg.2013.00279/ "Measuring belief in conspiracy theories: the generic conspiracist beliefs scale") but for now, the important thing to know is that a group of folks from Goldsmiths University of London put together a scale for measuring belief in conspiracy theories which they call the Generic Conspiracist Beliefs Scale. The dataset we're looking at has, for each of the testers, their responses to the fifteen questions on the scale and lots of other information about them (religion, handedness, level of education, etc.) You can redownload the data from [here](http://personality-testing.info/_rawdata/ "Raw Data") if you want to play with it yourself later.

```r
library(dplyr)
gcbs_raw <- read.csv('./GCBS/data.csv')
slice(gcbs_raw,1:5)
colnames(gcbs_raw)
```

<a id="reshape"></a>
### Reshaping Data

There's a whole lot here, right? My goodness. Let's make ourselves a more pliable data table, shall we? First, let's build a table out of the variables we're not going to change:

```r
gcbs <- gcbs_raw[c("education","age","familysize","testelapse",
                   "TIPI1","TIPI2","TIPI3","TIPI4","TIPI5","TIPI6","TIPI7","TIPI8",
                   "TIPI9","TIPI10")]
slice(gcbs,1:5)
```

If you're wondering what this TIPI is, by the way, it's a ten item personality inventory. Participants answer questions about themsevles on a scale of 1 (disagree strongly) to 7 (agree strongly).

TIPI1	Extraverted, enthusiastic.

TIPI2	Critical, quarrelsome.

TIPI3	Dependable, self-disciplined.

TIPI4	Anxious, easily upset.

TIPI5	Open to new experiences, complex.

TIPI6	Reserved, quiet.

TIPI7	Sympathetic, warm.

TIPI8	Disorganized, careless.

TIPI9	Calm, emotionally stable.

TIPI10	Conventional, uncreative.

Next, we're going to add in a bunch of variables that will need some manipulation to become useful. First, let's add in results from the word definition test:

```r
gcbs$wordliar <- ifelse((gcbs_raw$VCL6+gcbs_raw$VCL9+gcbs_raw$VCL12>0),TRUE,FALSE)
gcbs$knownwords <- (gcbs_raw$VCL1+gcbs_raw$VCL2+gcbs_raw$VCL3+gcbs_raw$VCL4+
                      gcbs_raw$VCL5+gcbs_raw$VCL7+gcbs_raw$VCL8+gcbs_raw$VCL10+
                      gcbs_raw$VCL11+gcbs_raw$VCL13+gcbs_raw$VCL14+gcbs_raw$VCL15+
                      gcbs_raw$VCL16)/13
slice(gcbs,1:5)
slice(filter(gcbs,wordliar==TRUE),1:5)
```

Now, let's put in some of the categorical variables that were numeric in the raw data. Note that instead of doing all of this complicated stuff, I could have just done as.factor(gcbs_raw$religion), but this way the output will be more readable later.

```r
gcbs$religion <- as.factor(c("Other","Agnostic.Atheist","Agnostic.Atheist","Other","Christian","Christian",
               "Christian","Christian","Other","Jewish","Muslim",
               "Other","Other")[as.factor(gcbs_raw$religion)])
gcbs$hand <- as.factor(c("other","right","left","both")[as.factor(gcbs_raw$hand)])
gcbs$gender <- as.factor(c("Other","Male","Female","Other")[as.factor(gcbs_raw$gender)])
gcbs$orientation <- as.factor(c("Other","Hetero","Bi","Homosexual","Other","Other")[as.factor(gcbs_raw$orientation)])
gcbs$race <- as.factor(c("Other","Asian","Arab","Black","Other","Other")[as.factor(gcbs_raw$race)])
gcbs$voted <- as.factor(c("Other","Yes","No"))[as.factor(gcbs_raw$voted)]
gcbs$urban <- as.factor(c("Other","Rural","Suburban","Urban"))[as.factor(gcbs_raw$urban)]
gcbs$married <- as.factor(c("Other","Never","Currently","Previously"))[as.factor(gcbs_raw$married)]
slice(gcbs,1:5)
```

Finally, let's add in the actual conspiracy scale. Rather than put in each of the questions, we should probably give it some sort of summary statistic. Also, here are the questions they used, if you're curious!
![GCBS questions](http://www.frontiersin.org/files/Articles/46573/fpsyg-04-00279-HTML/image_m/fpsyg-04-00279-t003.jpg)

```r
gcbs_raw$gm <- (gcbs_raw$Q1+gcbs_raw$Q6+gcbs_raw$Q11)
gcbs_raw$mg <- (gcbs_raw$Q2+gcbs_raw$Q7+gcbs_raw$Q12)
gcbs_raw$et <- (gcbs_raw$Q3+gcbs_raw$Q8+gcbs_raw$Q13)
gcbs_raw$pw <- (gcbs_raw$Q4+gcbs_raw$Q9+gcbs_raw$Q14)
gcbs_raw$ci <- (gcbs_raw$Q5+gcbs_raw$Q10+gcbs_raw$Q15)
gcbs_raw$overall <- (gcbs_raw$gm+gcbs_raw$mg+gcbs_raw$et+gcbs_raw$pw+gcbs_raw$ci)/15
gcbs$class <- as.factor(ifelse(gcbs_raw$overall>3.5,"High",
                               ifelse(gcbs_raw$overall>=2.5,"Med","Low")))
summary(gcbs$class)
colnames(gcbs)
slice(gcbs,1:5)
```

<a id="ttset"></a>
### Training and Test Sets

Looks like we have our data table! Now before we go on to do our trees and machines, let's prep a training and test set. (We've got enough data here that we can do that instead of cross validation.)

The "caret" package has a nice function for this called createDataPartition that pulls a stratified random sample of the data to make your training set.

```r
library(caret)
inTrain <- createDataPartition(y=gcbs$class,p=.75,list=FALSE)
inTrain[1:25]
trainSet <- gcbs[inTrain,]
testSet <- gcbs[-inTrain,]
#nrow(trainSet)
#nrow(testSet)
slice(trainSet,1:5)
```

<a id="dtree"></a>
## Decision Trees

We're going to use the rpart package to build our decision trees. Let's just hop right in and train a model!

```r
library(rpart)
rpart.model <- rpart(class~.,data=trainSet, method="class")
```

There are a few different ways we can inspect the model. Here are some!

```r
# Just print it
rpart.model
```

```r
# Print it, but fancier
summary(rpart.model)
```

```r
# Draw the tree
plot(rpart.model)
text(rpart.model)
```

```r
# Draw the tree, but better
library(rattle)
fancyRpartPlot(rpart.model)
```

Well, this looks fine, I guess. Let's see what happens when we try to run it on the training set.

```r
rpart.pred <- predict(rpart.model, testSet, type="class")
rpart.pred[1:5]
table(rpart.pred,testSet$class)
```

Looking down the diagonal of the prediction table, what do you notice? Yeah, it's awful. That's the sad truth about decision trees - for complex data, they really don't work very well. Maybe it's time to give their bigger, badder cousin a try! But first...

We might be able to improve our predictions a bit if we change around some parameters. What parameters can we work with here?

weights: we could weight the medium cases a little more strongly, since it seems to be uninterested in classifying cases as medium. This feels a bit dodgy, so I'd hold off on doing it.

minsplit: minimum number of observations that must exist in a node in order for a split to be attempted. What are some benefits and drawbacks of raising or lowering this? (default=20)

minbucket: minimum number of observations in any leaf node. Again, benefits and drawbacks of raising or lowering this? (default = round(minsplit/3), so 6 or 7)

cp: complexity parameter. Any split that doesn't decrease the overall lack of fit by a factor of cp is not attempted. (default = 0.01)

maxdepth: maximum depth of any node of the final tree, with root=0. Maxes out at 30, which is also the default.

split: splitting index, which can be gini (default) or information. A reminder that Gini impurity is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it were randomly labeled according to the distribution of labels in the subset (thanks, Wikipedia), whereas information gain measures how much entropy is reduced by the split.

```r
rpart.model <- rpart(class~.,data=trainSet, method="class",minsplit=5,parms = list(split = "information"))
fancyRpartPlot(rpart.model)
rpart.pred <- predict(rpart.model, testSet, type="class")
table(rpart.pred,testSet$class)
```

The e1071 package we'll use later for SVM classification has a function that helps tune parameters by running a grid search. It takes a bit of time and I'm not always thrilled with the output, but here's an example if you want it.

```r
tuned.rpart <- tune.rpart(class~., data = gcbs, minsplit = c(5,10,15))
summary(tuned.rpart)
plot(tuned.rpart)
```

<a id="forest"></a>
## Random Forests

Well, let's just jump right in here:

```r
library(randomForest)
rf.model <- randomForest(class~.,data=trainSet)
rf.model
```

This is also pretty awful, and it's just the training set. At this point, it's starting to look like our arbitrary splitting of the class into "high, medium, low" was poorly done. Let's try making it a binary variable instead, rebuild our datasets (remember, we want to keep the stratification) and try again.

```r
gcbs$class <- as.factor(ifelse(gcbs_raw$overall>2.75,"High","Low"))
inTrain <- createDataPartition(y=gcbs$class,p=.75,list=FALSE)
trainSet <- gcbs[inTrain,]
testSet <- gcbs[-inTrain,]
slice(trainSet,1:5)
```

```r
rf.model <- randomForest(class~.,data=trainSet)
rf.model
#summary(rf.model)
```

Well, maybe this isn't the greatest data set, then, but let's keep rolling. We might want to be able to see what variables are the most important in our model. Good news - this is really easy! We can either list it or plot it.

```r
#rf.model$importance
varImpPlot(rf.model)
```

To apply the model to our training set, we can actually use pretty similar code to what we used back in the decision tree section:

```r
rf.pred <- predict(rf.model, newdata=testSet, type="prob")
rf.pred[1:5,]
```

Cool! Now we can evaluate our performance. We'll make an ROC curve and then check the AUC.

```r
library(ROCR)
rf.prediction <- prediction(rf.pred[,2],testSet$class)
#slotNames(rf.prediction)
auc.perf <- performance(rf.prediction, measure="auc")
print(paste("AUC:",auc.perf@y.values))
rf.perf <- performance(rf.prediction,"tpr","fpr")
plot(rf.perf, main="ROC Curve for RF")
abline(0,1,lty=2)
```

Still a bit garbagey. Let's mess with some parameters. Here are some options: 
 
ntree: Number of trees to grow. Why might we want this to be larger or smaller? (default=500)

mtry: # of variable randomly sampled as candidates at each split (default=sqrt(# variables))

nodesize: minimum size of terminal nodes (default=1)

maxnodes: max number of terminal nodes a tree can have (default=max possible under nodesize)

There's a function called tuneRF that helps you choose the optimal value for mtry, but I haven't been super impressed with it and this is my class so we're going to crowdsource instead.

```r
rf.model <- randomForest(class~.,data=trainSet,mtry=10, nodesize=2)
varImpPlot(rf.model)
rf.pred <- predict(rf.model, newdata=testSet, type="prob")
rf.prediction <- prediction(rf.pred[,2],testSet$class)
#slotNames(rf.prediction)
auc.perf <- performance(rf.prediction, measure="auc")
print(paste("AUC:",auc.perf@y.values))
rf.perf <- performance(rf.prediction,"tpr","fpr")
plot(rf.perf, main="ROC Curve for RF")
abline(0,1,lty=2)
```

```r
# In case you want to play with it
tune.rf <- tuneRF(x=gcbs[,-ncol(gcbs)],y=gcbs$class)
tune.rf
```

### Briefly: Kernels!?

Kernels are kind of a tough thing to wrap one's head around, so let's quickly review. The basic idea is that a dataset that's too noisy to be separated in its own dimensions might be easier to separate if you map them to a higher dimension, like so:

<img src="./nonlinear.png">

I found an excellent example of the very popular radial basis function kernel [here](https://rstudio-pubs-static.s3.amazonaws.com/22171_ac34b133d26c4affb178031a9407d1ce.html#support-vector-machines "SVM Kernels!"), which I now present to you:

```r
# Generate some test data
set.seed (1)
x <- matrix(rnorm(200*2), ncol=2)
x[1:100,]=x[1:100,]+2
x[101:150,]=x[101:150,]-2
y <- c(rep(1,150),rep(2,50))
dat <- data.frame(x=x,y=as.factor(y))

plot(x, col=y)
```

```r
library(e1071)
# Create model and view results

train <- sample(200, 100)

svm.fit <- svm(y ~., data=dat[train,], kernel='radial', gamma=1, cost=1)

plot(svm.fit, dat[train,])

yhat <- predict(svm.fit, dat[-train,])
confusionMatrix(yhat, dat[-train,'y'])
```

As you can see, this kernel projects a sort of Gaussian deal - the closer to the center a point is the more likely it is to be in class 2. A more 3D view of what it's doing: <img src="http://i2.wp.com/blog.hackerearth.com/wp-content/uploads/2017/02/kernel.png"> This is very popular and is the default kernel in R. Here are some other kernels you might encounter!

The polynomial kernel, of varying degrees! <img src="https://openi.nlm.nih.gov/imgs/512/29/2547983/PMC2547983_pcbi.1000173.g006.png">

Or the sigmoid kernel, which is apparently too ugly for anyone to make a good graphic of! But mostly you'll see the RBF kernel. Unless your dataset is as simple as the one above, it's difficult to tell right off which kernel you'll want, so...experiment! And that's all I'm going to say about that.

<a id="old"></a>
### Back to the old dataset!

Let's jump right in. For reasons I don't quite understand, SVM makes much uglier curves if you give it a class rather than a regression, so here's a fake regression.

```r
trainSet$class <- ifelse(trainSet$class=="High",1,0)
testSet$class <- ifelse(testSet$class=="High",1,0)
```

```r
library(e1071)
svm.model <- svm(class~., data=trainSet, probability=TRUE)
svm.pred <- predict(svm.model, testSet, probability=TRUE,decision.values=TRUE)
svm.prediction <- prediction(attr(svm.pred,"decision.values"), testSet$class)
svm.perf <- performance(svm.prediction, "tpr", "fpr")
plot(svm.perf, main='ROC Curve for SVM Classifier')
abline(0,1,lty=2)
auc.perf <- performance(svm.prediction, measure="auc")
auc.perf@y.values
```

This is...okay, but not great. Parameters time!

kernel: as we've seen, the options are linear, polynomial, radial basis and sigmoid (default="radial")

gamma: the radius of influence of a support vector (low gamma -> far influence, high gamma -> local influence) (default=1/(data dimension))

cost: cost of a constraints violation (default=1)

Here's an example of the effect gamma and C changes may have on a classification

<img src="http://scikit-learn.org/stable/_images/sphx_glr_plot_rbf_parameters_001.png">

...and the validation accuracy. Where's the overfitting and where's the underfitting?

<img src="http://scikit-learn.org/stable/_images/sphx_glr_plot_rbf_parameters_002.png">

Again, like I said, there's also the possibility of doing a grid search for optimal parameters with the tune method. However, for SVM, that takes at least a half hour, and it isn't really worth it for now.

```r
svm.tune <- tune(svm,class~., data = trainSet, type="eps-regression", 
                 probability=TRUE ,ranges=list(cost=2^(-1:1), gamma=2^(1:4))) 
svm.tune
```

```r
svm.model <- svm(class~., data=trainSet, probability=TRUE, kernel="radial", cost=10, gamma=0.01)
svm.pred <- predict(svm.model, testSet, probability=TRUE,decision.values=TRUE)
svm.prediction <- prediction(attr(svm.pred,"decision.values"), testSet$class)
svm.perf <- performance(svm.prediction, "tpr", "fpr")
plot(svm.perf, main='ROC Curve for SVM Classifier')
abline(0,1,lty=2)
auc.perf <- performance(svm.prediction, measure="auc")
auc.perf@y.values
```

### It's not you, it's me (me being the terrible dataset)

Okay, just because I want to prove to you that SVM actually can work when you have a reasonable dataset, I present to you its results on the [UCI Iris dataset](https://archive.ics.uci.edu/ml/datasets/Iris "UCI Iris dataset") (probably the most famous dataset in all of ML).

```r
iris <- read.csv("./iris.csv", header=TRUE)

iris <- filter(iris,!(iris$class=="Iris-setosa"))
iris$class <- ifelse(iris$class=="Iris-virginica",1,0)
iris.inTrain <- createDataPartition(y=iris$class,p=.75,list=FALSE)
iris.trainSet <- iris[iris.inTrain,]
iris.testSet <- iris[-iris.inTrain,]

iris.model <- svm(class~., data=iris.trainSet, probability=TRUE)
iris.pred <- predict(iris.model, iris.testSet, probability=TRUE,decision.values=TRUE)
iris.prediction <- prediction(attr(iris.pred,"decision.values"), iris.testSet$class)
iris.perf <- performance(iris.prediction,"tpr","fpr")
plot(iris.perf,main="I told you SVM can work!")
abline(0,1,lty=2)
auc.perf <- performance(iris.prediction, measure="auc")
auc.perf@y.values
```
